# GRNgene simulations

## Ecoli size equivalent network: 1565 genes

```python
# Set parameters

nbgene = 1565
autoRG = 0
duoRG = 0
an = 5

G = GRN.random_grn(genesNb=nbgene,
              autoRG=autoRG,
              duoRG=duoRG,
              an=an
              )

# Load the adjacency matrix as binary

grgene_adj_mx = G['AdjMatrice'].copy()
grgene_adj_mx[grgene_adj_mx!=0] = 1.
```

```python
# Plot the degree distribution
plot_grn_degree(grgene_adj_mx, plot_network=True)
```

```python
# Set parameters
for an in [2, 10, 100, 500, 1000]:
    nbgene = 1565
    autoRG = 0
    duoRG = 0

    G = GRN.random_grn(genesNb=nbgene,
                autoRG=autoRG,
                duoRG=duoRG,
                an=an
                )

    # Load the adjacency matrix as binary

    grgene_adj_mx = G['AdjMatrice'].copy()
    grgene_adj_mx[grgene_adj_mx!=0] = 1.
    # Plot the degree distribution
    plot_grn_degree(grgene_adj_mx)
```
